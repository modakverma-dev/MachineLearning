```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def MSE(x,y,m,b):
    mse=0
    x = np.array(x)
    y = np.array(y)
    mse = ((y-(m*x+b))**2).mean()
    return mse 
```

```python
def gradient_descent(x,y,epochs,step):
    x=np.array(x)
    y=np.array(y)
    m=0
    b=0
    for i in range(epochs):
        delta = y-(m*x+b)
        del_m = -2*((x*delta).mean())
        del_b = -2*((delta).mean())
        m-=step*del_m
        b-=step*del_b
        print(f"Epoch {i}, error: {MSE(x,y,m,b)}")
    return m,b
```

```python
x=[1,2,3,4,5]
y=[5,7,9,11,13]
epochs=1
learning_rate=0.05
m,b = gradient_descent(x,y,epochs,learning_rate)
print (m,b)
```

```python
plt.scatter(x,y,marker="x",color="red")
x1=np.array(x)
y1=m*x1+b
plt.plot(x1,y1)
plt.plot(x,y)
plt.grid()
```
